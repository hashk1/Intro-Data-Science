```python
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.formula.api as smf
%matplotlib inline
plt.style.use("ggplot")
```

### CSVファイルを読みこむ

```python
# CSVファイルを読みこむ
ad_data = pd.read_csv("../Data/ad_result.csv")
ad_data.head()
```

### TVの広告費と新規ユーザ数の散布図をつくる

```python
ad_data.plot(x = "tvcm", y = "install", kind = "scatter")
plt.xlabel("TVの広告費")
plt.ylabel("新規インストール")
```

### 雑誌の広告費と新規ユーザ数の散布図をつくる

```python
ad_data.plot(x = "magazine", y = "install", kind = "scatter")
plt.xlabel("雑誌の広告費")
plt.ylabel("新規インストール")
```

### 回帰分析の実行

```python
# 回帰分析の実行
model = smf.ols(formula = 'install ~ tvcm + magazine', data = ad_data)
result = model.fit()
result.params
```

###  回帰分析の結果の解釈

```python
result.summary()
```
